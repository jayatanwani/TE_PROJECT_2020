```python
import pandas as pd
```

```python
data=pd.read_csv("2016.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
states=data['State Name'].tolist()
states
```

```python
labelencoder=LabelEncoder()
labelencoder.fit(states)
labels=labelencoder.transform(states)
data['State Name']=pd.Series(labels)
data
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
#data['Station Name']=le.fit_transform(data['Station Name'])
data['temp_min']=le.fit_transform(data['temp_min'])
data['oxy_min']=le.fit_transform(data['oxy_min'])
data['temp_min']=le.fit_transform(data['ph_min'])
data['temp_max']=le.fit_transform(data['temp_max'])
data['oxy_max']=le.fit_transform(data['oxy_max'])
data['ph_max']=le.fit_transform(data['ph_max'])
data
```

```python
cols = [col for col in data.columns if col not in ['bact','Station Name']]
cols
```

```python
x=data[cols]
y=data['bact']
x
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(x,y, test_size = 0.20, random_state = 10)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
gnb = GaussianNB()
pred = gnb.fit(data_train, target_train).predict(data_test)
print("Naive-Bayes accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
# Naive Bayes
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(data_train, target_train)
 
y_pred = classifier.predict(data_test)
 
# Summary of the predictions made by the classifier
print(classification_report(target_test, y_pred))
print(confusion_matrix(target_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,target_test))
```

```python
from yellowbrick.classifier import ClassificationReport
# Instantiate the classification model and visualizer
visualizer = ClassificationReport(gnb, classes=['low bact growth','high bact growth'])
visualizer.fit(data_train, target_train) # Fit the training data to the visualizer
visualizer.score(data_test, target_test) # Evaluate the model on the test data
g = visualizer.poof() # Draw/show/poof the data
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
svc_model = LinearSVC(random_state=0)
pred = svc_model.fit(data_train, target_train).predict(data_test)
print("LinearSVC accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
# Linear SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.svm import LinearSVC
classifier = LinearSVC(random_state=0)
classifier.fit(data_train, target_train)
 
y_pred = classifier.predict(data_test)
 
# Summary of the predictions made by the classifier
print(classification_report(target_test, y_pred))
print(confusion_matrix(target_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,target_test))
```

```python
from yellowbrick.classifier import ClassificationReport
# Instantiate the classification model and visualizer
visualizer = ClassificationReport(svc_model, classes=['bact low','bact high'])
visualizer.fit(data_train, target_train) # Fit the training data to the visualizer
visualizer.score(data_test, target_test) # Evaluate the model on the test data
g = visualizer.poof() # Draw/show/poof the data
```

```python
#import necessary modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
#create object of the lassifier
neigh = KNeighborsClassifier(n_neighbors=2)
#Train the algorithm
neigh.fit(data_train, target_train)
# predict the response
pred = neigh.predict(data_test)
# evaluate accuracy
print ("KNeighbors accuracy score : ",accuracy_score(target_test, pred))
```

```python
# K neighbors
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(data_train, target_train)
 
y_pred = classifier.predict(data_test)
 
# Summary of the predictions made by the classifier
print(classification_report(target_test, y_pred))
print(confusion_matrix(target_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,target_test))
```

```python
from yellowbrick.classifier import ClassificationReport
# Instantiate the classification model and visualizer
visualizer = ClassificationReport(neigh, classes=['high bact','low bact'])
visualizer.fit(data_train, target_train) # Fit the training data to the visualizer
visualizer.score(data_test, target_test) # Evaluate the model on the test data
g = visualizer.poof() # Draw/show/poof the data
```
